```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
C=QuantumCircuit(3,3)
```

```python
C.draw(output='mpl')
```

```python
C.x(0)
C.barrier()
C.draw(output='mpl')
```

```python
C.x(1)
C.cx(1,2)
C.draw(output='mpl')
```

```python
C.cx(0,1)
C.h(0)
C.draw(output='mpl')
```

```python
C.barrier()
C.measure([0,1],[0,1])
C.draw(output='mpl')
```

```python
C.barrier()
C.cx(1,2)
C.cz(0,2)
C.draw(output='mpl')
```

```python
C.measure(2,2)
simulator=Aer.get_backend('qasm_simulator')
result=execute(C,backend=simulator,shots=1024).result()
counts=result.get_counts()
plot_histogram(counts)
```
